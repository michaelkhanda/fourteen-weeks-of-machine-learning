```python
import matplotlib.pyplot as plt
import mglearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

#make the models
log_reg_model = LogisticRegression(max_iter=1000000).fit(X_train,y_train)
svc_model = LinearSVC(max_iter=1000000).fit(X_train,y_train)

print("The training accuracy is :{}".format(log_reg_model.score(X_train, y_train)))
print("The training accuracy is :{}".format(svc_model.score(X_train, y_train)))
```

```python
fig,axes = plt.subplots(1,2)

for model ,ax in zip([LinearSVC(max_iter=1000000), LogisticRegression(max_iter=1000000)], axes):
    clf = model.fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X, ax=ax)
    
    mglearn.discrete_scatter(X[:,0], X[:,1],y, ax=ax)
axes[1].legend()
axes[0].legend()
```

```python
#logistic Regression for Cancer dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)

log_reg1 =LogisticRegression (max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(log_reg1.score(X_train, y_train)))
print("Testing set score: {:.2f}".format(log_reg1.score(X_test, y_test)))
```

```python
log_reg10 = LogisticRegression(C=10, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(log_reg10.score(X_train, y_train)))
print("Testing set score: {:.2f}".format(log_reg10.score(X_test, y_test)))
```

```python
log_reg0001 = LogisticRegression(C=0.001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(log_reg0001.score(X_train, y_train)))
print("Testing set score: {:.2f}".format(log_reg0001.score(X_test, y_test)))
```

```python
plt.plot(log_reg1.coef_.T, 's', label = "C=1")
plt.plot(log_reg10.coef_.T, 's', label = "C=10")
plt.plot(log_reg0001.coef_.T, 's', label = "C=0.001")
plt.legend()
```
