```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y = np.array([60, 80, 70, 100, 150])

lr = LinearRegression()
lr.fit(X, y)

plt.scatter(X, y)
#plt.plot(X, lr.coef_*X+lr.intercept_, color='r')
plt.plot(X, lr.predict(X), color='g')  # Use predict() to get the predicted y values
plt.xlabel("X")
plt.ylabel("y")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y = np.array([60, 80, 70, 100, 150])

lr = LinearRegression()
lr.fit(X, y)

a = lr.coef_[0]
b = lr.intercept_

plt.scatter(X, y)
plt.plot(X, a * X + b, color='r')
plt.xlabel("X")
plt.ylabel("y")

plt.show()
```
